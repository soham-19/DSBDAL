**Perform the following operations on iris dataset**

2. Create a box plot and histogram for each feature in the dataset.
3. Compare distributions and identify outliers.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# List down features

```python
df = pd.read_csv("Iris.csv")
df.head(2)
```

```python
df.boxplot(['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
```

```python
for col in df.columns[1:]:
    sns.histplot(data=df, x=col,kde=True)
    plt.show()

for col in df.columns[1:]:
    sns.boxplot(data=df, y=col)
    plt.show()
```

```python
for col in df.columns[1:-1]:
    print(col, end = "\t")
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lb = q1 - 1.25*iqr
    ub = q3 + 1.25*iqr
    print(f"q1 : {q1}, q3 : {q3}, iqr : {iqr}, lb : {lb}, ub : {ub}\n")

    outliers = df[(df[col] > ub) | (df[col] < lb)]
    print(outliers)
    df.boxplot([col])
    plt.show()
    print("\n\n")
```
