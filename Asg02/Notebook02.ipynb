# 4.Handle outlier using any technique

```python
#calculating q1,q2,IQR, lowerbouundry and upperboundry
q1 = df['FE'].quantile(0.25)
q3 = df['FE'].quantile(0.75)
iqr = q3 - q1    #Interquartile range
lower_boundry = q1 - (1.5*iqr)
upper_boundry = q3 + (1.5*iqr)
print("IQR", iqr, "Lower and upper boundry ",lower_boundry,upper_boundry )
```

```python
sns.boxplot(df['FE'])
plt.show()
```

```python
#finding outliers having value <  lower_boundry and > upper_boundry
outliers = df[(df['FE'] < lower_boundry) | (df['FE'] > upper_boundry)]
```

```python
outliers
```

```python
# drop rows containing outliers
df = df.drop(outliers.index)
```

```python
#find oulier in tecomp
from scipy import stats
z = np.abs(stats.zscore(df['BE']))
print(z)
```

```python
#remove outlier having z score > 2
threshold_z = 2

outlier_indices = np.where(z > threshold_z)[0]
no_outliers = df.drop(outlier_indices)
print("Original DataFrame Shape:", df.shape)
print("DataFrame Shape after Removing Outliers:", no_outliers.shape)
no_outliers
```

```python
# Define the range of the 'FEmarks' column
min_val = df['FE'].min()
max_val = df['FE'].max()

# Perform Min-Max normalization
min_max_normalized = (df['FE'] - min_val) / (max_val - min_val)

# Define a threshold for outliers in the normalized values
threshold_min_max = 0.9  # for example, considering values greater than 0.9 as outliers

# Find outliers based on Min-Max normalization
outlier_indices = np.where(min_max_normalized > threshold_min_max)[0]

# Remove outliers from the DataFrame
no_outliers = df.drop(outlier_indices)

# Print information about the original and modified DataFrame
print("Original DataFrame Shape:", df.shape)
print("DataFrame Shape after Removing Outliers:", no_outliers.shape)
print("Original DataFrame with Outliers:")
print(df.tail())
print("\nDataFrame without Outliers:")
print(no_outliers.tail())
```

```python
df = pd.read_csv("academicperformance.csv")
print("Before " , df.shape)
df.dropna(inplace = True)
print("After " , df.shape)
# Define the lower and upper thresholds for Winsorizing
lower_threshold = 5  # Example lower threshold (5th percentile)
upper_threshold = 95  # Example upper threshold (95th percentile)

# Winsorize the 'Femarks' column
df['FE_winsorized'] = df['FE'].clip(df['SE'].quantile(lower_threshold/100), df['FE'].quantile(upper_threshold/100))

# Winsorize the 'SEMarks' column
df['SE_winsorized'] = df['SE'].clip(df['SE'].quantile(lower_threshold/100), df['SE'].quantile(upper_threshold/100))

# Display the DataFrame with Winsorized values
# print(df)
```
